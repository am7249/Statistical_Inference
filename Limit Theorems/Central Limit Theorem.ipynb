# Central Limit Theorem

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

plt.style.use('ggplot')
```

```python
print(plt.style.available)
plt.style.use('seaborn-notebook')
```

```python
class CLT:
    def __init__(self, mean, no_means, sample):
        self.mean = mean
        self.s_no = no_means
        self.sample_size = sample
    
    def poisson(self):
        s_means = []
        for i in range(self.s_no):
            X = np.random.poisson(self.mean, size = self.sample_size)
            Z = np.divide((np.mean(X) - self.mean), np.std(X))
            s_means.append(Z)
        return s_means
    
    def pareto(self):
        s_means = []
        for i in range(self.s_no):
            X = np.random.pareto(self.mean, size = self.sample_size)
            Z = np.divide((np.mean(X) - self.mean), np.std(X))
            s_means.append(Z)
        return s_means
```

```python
clt1 = CLT(19,100000,100)
p1 = clt1.pareto()
```

```python
clt = CLT(19,50000,100)
p = clt.pareto()
```

```python
clt2 = CLT(19,30000,100)
p2 = clt2.pareto()
```

```python
plt.figure(figsize=(20,10))
plt.hist(p,color="g",bins=100,alpha=0.5)
plt.hist(p2,color='r',bins = 100,alpha=0.3)
plt.hist(p1,color="#3F5D7D", bins=100,alpha=0.7)
plt.show()
```
